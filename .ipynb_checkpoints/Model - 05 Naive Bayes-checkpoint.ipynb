### Import Packages and Reading of Data

```python
import time
import warnings
import pickle

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, LogisticRegression
from sklearn import metrics
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, cross_val_score, RepeatedKFold
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.inspection import permutation_importance

pd.set_option("display.max_columns", None)
warnings.filterwarnings("ignore")
```

```python
# Reading of Excel
dataset = pd.read_csv('cleaned_others_dataset.csv')
#dataset = dataset.drop(columns=["protected", "is_quoted_retweets"])
```

```python
dataset
```

### Feature Engineering

```python
x_train, x_test, y_train, y_test = train_test_split(dataset.drop(columns=["isFraud"]), dataset["isFraud"], test_size = 0.2, random_state=101)
print("Train set:", x_train.shape, y_train.shape)
print("Test set:", x_test.shape, y_test.shape)
```

```python
scaler = StandardScaler()
fitted = scaler.fit(x_train)
X_train_fitted = fitted.transform(x_train)
X_test_fitted = fitted.transform(x_test)
```

```python
x_train = pd.DataFrame(X_train_fitted, columns = x_train.columns)
x_test = pd.DataFrame(X_test_fitted, columns = x_test.columns)
y_train = pd.DataFrame(y_train, columns = ["isFraud"])
y_test = pd.DataFrame(y_test, columns = ["isFraud"])
```

```python
x_bin_train = x_train[["verified", 
               "location", 
               "un_special_char", 
               "un_uppercase", 
               "name_special_char", 
               "name_uppercase", 
               "des_external_links", 
               "has_description"]]
x_bin_test = x_test[["verified", 
               "location", 
               "un_special_char", 
               "un_uppercase", 
               "name_special_char", 
               "name_uppercase", 
               "des_external_links", 
               "has_description"]]
x_num_train = x_train[["followers_count", 
                 "following_count", 
                 "tweet_count", 
                 "un_no_of_char", 
                 "name_no_of_char", 
                 "des_no_of_usertags", 
                 "des_no_of_hashtags", 
                 "account_age_in_days", 
                 "average_tweets_per_day"]]
x_num_test = x_test[["followers_count", 
                 "following_count", 
                 "tweet_count", 
                 "un_no_of_char", 
                 "name_no_of_char", 
                 "des_no_of_usertags", 
                 "des_no_of_hashtags", 
                 "account_age_in_days", 
                 "average_tweets_per_day"]]
```

# Naive Bayes

* Error Metrics
* Plot for Feature Importance

### Grid Search

#### GaussianNB

```python
parameters = {
    'var_smoothing': np.logspace(0,-9, num=200) 
}
    
grid_search = GridSearchCV(
    estimator=GaussianNB(),
    param_grid=parameters,
    scoring = 'f1',
    n_jobs = -1,
    cv = 5,
    verbose=True
)
```

```python
%%time

grid_search.fit(x_num_train, y_train)
```

```python
print(grid_search.best_estimator_)
model_gnb = grid_search.best_estimator_
y_pred = model_gnb.fit(x_num_train, y_train).predict(x_num_test)
lg_loss = metrics.log_loss(y_test, y_pred)
print(f'Log Loss: {lg_loss}')
roc_auc = metrics.roc_auc_score(y_test, y_pred)
print(f'ROC AUC: {roc_auc}')
precision = metrics.precision_score(y_test, y_pred)
print(f'Precision: {precision}')
recall = metrics.recall_score(y_test, y_pred)
print(f'Recall: {recall}')
f1_score = metrics.f1_score(y_test, y_pred)
print(f'F1-score: {f1_score}')
```

### Randomized Search

#### GaussianNB

```python
parameters = {
    'var_smoothing': np.logspace(0,-9, num=200) 
}
    
random_search = RandomizedSearchCV(
    estimator=GaussianNB(),
    param_distributions=parameters,
    scoring = 'f1',
    n_jobs=-1,
    cv=5,
    verbose=10,
    n_iter=50,
    random_state=123
)
```

```python
%%time

random_search.fit(x_num_train, y_train)
```

```python
print(random_search.best_estimator_)
model_gnb = random_search.best_estimator_
y_pred = model_gnb.fit(x_num_train, y_train).predict(x_num_test)
lg_loss = metrics.log_loss(y_test, y_pred)
print(f'Log Loss: {lg_loss}')
roc_auc = metrics.roc_auc_score(y_test, y_pred)
print(f'ROC AUC: {roc_auc}')
precision = metrics.precision_score(y_test, y_pred)
print(f'Precision: {precision}')
recall = metrics.recall_score(y_test, y_pred)
print(f'Recall: {recall}')
f1_score = metrics.f1_score(y_test, y_pred)
print(f'F1-score: {f1_score}')
```

### Feature Importance
https://stackoverflow.com/questions/62933365/how-to-get-the-feature-importance-in-gaussian-naive-bayes

```python
optim_model_num = GaussianNB(var_smoothing=0.07401959996915641)
optim_model_num.fit(x_num_train, y_train)
imps = permutation_importance(optim_model_num, x_num_test, y_test)
feature = pd.DataFrame(imps.importances_mean, x_num_train.columns.T, columns=['Feature Importance']).sort_values(by='Feature Importance', ascending=False)
feature
```

```python
optim_model_bin = BernoulliNB(alpha=0)
optim_model_bin.fit(x_bin_train, y_train)
#print(optim_model_bin.class_log_prior_)
imps = permutation_importance(optim_model_bin, x_bin_test, y_test)
feature = pd.DataFrame(imps.importances_mean, x_bin_train.columns.T, columns=['Feature Importance']).sort_values(by='Feature Importance', ascending=False)
feature
```
