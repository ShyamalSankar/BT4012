```python
import pandas as pd
import seaborn as sns
data = pd.read_csv('cleaned_tweets_dataset.csv')
data
```

```python
import re
def hashtags_in_tweet(df):
    df['hashtags'] =''
    for word in range(len(df)):
        tweet = df["uncleaned_text"][word]
        tweet_clean = df['text'][word]
        tweet.replace('\n', " ")
        word_list = tweet.split(' ')
        word_list2 = tweet_clean.split(' ')
        regex = "#(\w+)"
        hashtag_list = re.findall(regex, tweet)
        for i in range(len(word_list2.copy())):
            hashtag_list_lower = [item.lower() for item in hashtag_list]
            if word_list2.copy()[i] in hashtag_list_lower:
                word_list2[i] = word_list2.copy()[i]
        new_tweet = " ".join(word_list2)
        df['text'][word] = new_tweet
        df["hashtags"][word] = hashtag_list
    return df

data = hashtags_in_tweet(data)
data
```

```python
import language_tool_python
tool = language_tool_python.LanguageTool('en-US')
data['no_of_grammar_errors'] = ''
for word in range(len(data)):
    cap = data["text"][word]
    word_list = cap.split(' ')
    data['no_of_grammar_errors'][word] = len(tool.check(cap))
data
```

```python
import ast
print(type(data['hashtags'][2]))
data['mentions_count'] = data['tags'].apply(lambda x: len(ast.literal_eval(x)))
data['hashtags_count'] = data['hashtags'].apply(lambda x: len(x))
data['emoji_count'] = data['emoji'].apply(lambda x: len(ast.literal_eval(x)))
human_tweets = data[data['account_type']== 'human']
bot_tweets = data[data['account_type']== 'bot']
data
data.to_csv("OldSet_FE.csv")
```
